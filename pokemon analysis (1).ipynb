#### Since the information about the height and weight value of pokemon is available on google, we'll fill the missing value using them

```python
height_weight_dict = {
    'Rattata': (0.3, 3.5),
    'Raticate': (0.7, 18.5),
    'Raichu': (0.8, 30.0),
    'Sandshrew': (0.6, 12.0),
    'Sandslash': (1.0, 29.5),
    'Vulpix': (0.6, 9.9),
    'Ninetales': (1.1, 19.9),
    'Diglett': (0.2, 0.8),
    'Dugtrio': (0.7, 33.3),
    'Meowth': (0.4, 4.2),
    'Persian': (1.0, 32.0),
    'Geodude': (0.4, 20.0),
    'Graveler': (1.0, 105.0),
    'Golem': (1.4, 300.0),
    'Grimer': (0.9, 30.0),
    'Muk': (1.2, 30.0),
    'Exeggutor': (2.0, 120.0),
    'Marowak': (1.0, 45.0),
    'Hoopa': (0.5, 9.0),
    'Lycanroc': (0.8, 25.0)
}

for index,row in df.iterrows():
    pokemon_names=row['name']
    if pokemon_names in height_weight_dict:
        height,weight=height_weight_dict[pokemon_names]
        df.at[index,'height_m']=height
        df.at[index,'weight_kg']=weight
```

```python
df[['name','type2']].loc[df['type2'].isna()]
```

### Strongest and Weakest type

```python
against_features = ['against_bug', 'against_dark', 'against_dragon', 'against_electric',
                    'against_fairy', 'against_fight', 'against_fire', 'against_flying',
                    'against_ghost', 'against_grass', 'against_ground', 'against_ice',
                    'against_normal', 'against_poison', 'against_psychic', 'against_rock',
                    'against_steel', 'against_water']
type_avg_damage = {}

for type_col in ['type1', 'type2']:
    for _, row in df.iterrows():
        pokemon_type = row[type_col]
        if pokemon_type not in type_avg_damage:
            type_avg_damage[pokemon_type] = [0] * len(against_features)
        
        for i, feature in enumerate(against_features):
            type_avg_damage[pokemon_type][i] += row[feature]

# Now calculate the average for each type
for type_key in type_avg_damage:
    total_pokemon = len(df[(df['type1'] == type_key) | (df['type2'] == type_key)])
    for i in range(len(against_features)):
        type_avg_damage[type_key][i] /= total_pokemon
type_avg_damage_summary = {type_key: sum(damages) / len(damages) for type_key, damages in type_avg_damage.items()}
strongest_type = min(type_avg_damage_summary, key=type_avg_damage_summary.get)
weakest_type = max(type_avg_damage_summary, key=type_avg_damage_summary.get)
```

```python
avg_damage_df = pd.DataFrame.from_dict(type_avg_damage_summary, orient='index', columns=['avg_damage']).reset_index()
avg_damage_df.rename(columns={'index': 'type'}, inplace=True)
plt.figure(figsize=(14, 8))
sns.barplot(x='avg_damage', y='type', data=avg_damage_df, palette='viridis')
plt.title('Average Damage Multiplier Against Different Types')
plt.xlabel('Average Damage Multiplier')
plt.ylabel('Type')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

- The strongest type is steel with an average damage multiplier of 0.88
- The weakest type is ice with an average damage multiplier of 1.20
